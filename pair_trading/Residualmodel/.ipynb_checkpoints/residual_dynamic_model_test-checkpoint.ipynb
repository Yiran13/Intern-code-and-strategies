```python
from datetime import datetime
from vnpy.app.portfolio_strategy import BacktestingEngine
from vnpy.trader.constant import Interval
from residualmodel_dynamic import DynamicResidualModelStrategy
import pandas as pd
import numpy as np
```

```python
engine = BacktestingEngine()

engine.set_parameters(
    vt_symbols=["HC888.SHFE", 'RB888.SHFE'],
    interval=Interval.MINUTE,
    start=datetime(2019, 3, 1 ),
    end=datetime(2019, 10, 1),
    rates={"HC888.SHFE": 5/10000, "RB888.SHFE": 5/10000},
    slippages={"HC888.SHFE": 1, "RB888.SHFE": 0.5},
    sizes={"HC888.SHFE":10, "RB888.SHFE":10},
    priceticks={"HC888.SHFE":2, "RB888.SHFE":1},
    capital=1_000_0,
    collection_names={"HC888.SHFE":"HC888", "RB888.SHFE":"RB888"}

)
engine.add_strategy(DynamicResidualModelStrategy, {'x_multiplier': 1.58,'y_multiplier': 2, 'intercept': 1328})
engine.load_data()
engine.run_backtesting()
df = engine.calculate_result()
engine.calculate_statistics()
engine.show_chart()
```

### 策略参数优化记录

对5个交易品种进行下面3个参数的优化

- `short_entry_multiplier`,`long_entry_multiplier` 绝对值相同，方向相反 [2,3,4,5,6]
- `difference_filter_num` [20,40,60,80],`difference_exit_num`设置为`difference_filter_num` 的1/2
- ` std_window` [30,60,90,120,150,180],`mean_window`是其1/2

```python
HC_RB_filtered = HC_RB_df[(HC_RB_df['sharpe_ratio']>1)&(HC_RB_df['max_ddpercent']>-20)]
```

```python
short_entry_multiplier_array = []
difference_filter_num_array = []
std_window_array = []
for index in range(HC_RB_filtered.shape[0]):
    param = eval(HC_RB_filtered['param'].iloc[index])
    short_entry_multiplier_array.append(param['short_entry_multiplier'])
    difference_filter_num_array.append(param['difference_filter_num'])
    std_window_array.append(param['std_window'])
```

```python
HC_RB_filtered.loc[:,'short_entry_multiplier'] =  short_entry_multiplier_array
HC_RB_filtered.loc[:,'difference_filter_num'] = difference_filter_num_array
HC_RB_filtered.loc[:,'std_window'] = std_window_array
del HC_RB_filtered['param']
```
